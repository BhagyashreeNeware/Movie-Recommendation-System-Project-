```python
a = "select * from movies"
a=sqldf(a)
print(a)
```

```python
b = "select avg(budget) from movies"
c = sqldf(b)
print(c)
```

```python
e = "SELECT title, popularity ,budget from movies order by popularity desc limit 10"
e = sqldf(e)
print(e)
```

```python
f = "SELECT title, AVG(vote_count) AS avg_vote FROM movies GROUP BY title ORDER BY avg_vote DESC"
f = sqldf(f)
print(f)
```

```python
f = "SELECT title,vote_count, AVG(vote_count) AS avg_vote FROM movies GROUP BY title ORDER BY avg_vote DESC"
f = sqldf(f)
print(f)
```

```python
g = "SELECT id, COUNT(*) AS budget FROM movies GROUP BY title ORDER BY id asc"
g = sqldf(g)
print(g)
```

```python
h = "select distinct(vote_average) from movies"
h = sqldf(h)
print(h)
```

```python
i = "SELECT id, AVG(vote_count) AS avg_vote FROM movies GROUP BY vote_average ORDER BY avg_vote DESC"
i = sqldf(i)
print(i)
```

```python
-- Most popular genres


-- Average ratings per genre
SELECT genre, AVG(rating) AS avg_rating
FROM movies
JOIN user_ratings ON movies.movie_id = user_ratings.movie_id
GROUP BY genre
ORDER BY avg_rating DESC;
```

```python
# -- Most popular genres
j = "SELECT genres, COUNT(*) AS num_ratings FROM movies GROUP BY genres ORDER BY num_ratings DESC"
j = sqldf(j)
print(j)
```

```python
#-- Average voting per genres
h = "SELECT genres, AVG(vote_count) AS avg_vote FROM movies GROUP BY genres ORDER BY avg_vote DESC"
h = sqldf(h)
print(h)
```
